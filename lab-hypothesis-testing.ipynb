- We posit that Pokemons of type Dragon have, on average, more HP stats than Grass. Choose the propper test and, with 5% significance, comment your findings.

```python
from scipy.stats import ttest_ind
import pandas as pd

def compare_hp_between_types(df, type_a='Dragon', type_b='Grass', alpha=0.05):
    # Filter Pokémon with the specific types in either Type 1 or Type 2
    group_a = df[(df['Type 1'] == type_a) | (df['Type 2'] == type_a)]['HP'].dropna()
    group_b = df[(df['Type 1'] == type_b) | (df['Type 2'] == type_b)]['HP'].dropna()

    # Perform Welch’s t-test (assumes unequal variances)
    t_stat, p_value = ttest_ind(group_a, group_b, equal_var=False)

    result = {
        'mean_HP_'+type_a: round(group_a.mean(), 2),
        'mean_HP_'+type_b: round(group_b.mean(), 2),
        't_statistic': round(t_stat, 4),
        'p_value': round(p_value, 4),
        'significant': p_value < alpha,
        'conclusion': 'Reject null: Dragon HP > Grass' if p_value < alpha else 'Fail to reject null: No significant difference'
    }
    return result

result = compare_hp_between_types(pokemon_df)
print(result)
```

- We posit that Legendary Pokemons have different stats (HP, Attack, Defense, Sp.Atk, Sp.Def, Speed) when comparing with Non-Legendary. Choose the propper test and, with 5% significance, comment your findings.


```python
import pandas as pd
from statsmodels.multivariate.manova import MANOVA
from scipy.stats import ttest_ind

def legendary_vs_nonlegendary_stats(df, alpha=0.05):
    # Select relevant stats and group label
    stats = ['HP', 'Attack', 'Defense', 'Sp. Atk', 'Sp. Def', 'Speed']
    df_filtered = df[stats + ['Legendary']].dropna()

    # MANOVA test
    formula = ' + '.join(stats) + ' ~ Legendary'
    manova = MANOVA.from_formula(formula, data=df_filtered)
    manova_result = manova.mv_test().results['Legendary']['stat']

    # Follow-up: t-tests per stat
    ttest_results = {}
    group_legend = df_filtered[df_filtered['Legendary'] == True]
    group_nonlegend = df_filtered[df_filtered['Legendary'] == False]

    for stat in stats:
        t_stat, p_value = ttest_ind(group_legend[stat], group_nonlegend[stat], equal_var=False)
        ttest_results[stat] = {
            'Legendary_mean': round(group_legend[stat].mean(), 2),
            'NonLegendary_mean': round(group_nonlegend[stat].mean(), 2),
            't_stat': round(t_stat, 4),
            'p_value': round(p_value, 4),
            'significant': p_value < alpha
        }

    return {
        'MANOVA': manova_result,
        't_tests': ttest_results
    }

results = legendary_vs_nonlegendary_stats(pokemon_df)
print(results['MANOVA'])  # Summary of multivariate test
print(results['t_tests'])  # Detailed univariate results
```

**We posit that houses close to either a school or a hospital are more expensive.**

- School coordinates (-118, 34)
- Hospital coordinates (-122, 37)

We consider a house (neighborhood) to be close to a school or hospital if the distance is lower than 0.50.

Hint:
- Write a function to calculate euclidean distance from each house (neighborhood) to the school and to the hospital.
- Divide your dataset into houses close and far from either a hospital or school.
- Choose the propper test and, with 5% significance, comment your findings.
 

```python
import numpy as np

def add_distance_columns(df):
    school_coords = (-118, 34)
    hospital_coords = (-122, 37)
    
    df['dist_to_school'] = np.sqrt((df['longitude'] - school_coords[0])**2 + (df['latitude'] - school_coords[1])**2)
    df['dist_to_hospital'] = np.sqrt((df['longitude'] - hospital_coords[0])**2 + (df['latitude'] - hospital_coords[1])**2)
    
    df['is_close'] = (df['dist_to_school'] < 0.5) | (df['dist_to_hospital'] < 0.5)
    return df
```

```python
housing_df = add_distance_columns(housing_df)
```

```python
from scipy.stats import ttest_ind

def test_house_value_by_proximity(df, alpha=0.05):
    close_group = df[df['is_close'] == True]['median_house_value'].dropna()
    far_group = df[df['is_close'] == False]['median_house_value'].dropna()
    
    t_stat, p_value = ttest_ind(close_group, far_group, equal_var=False)

    return {
        'mean_close': round(close_group.mean(), 2),
        'mean_far': round(far_group.mean(), 2),
        't_statistic': round(t_stat, 4),
        'p_value': round(p_value, 4),
        'significant': p_value < alpha,
        'conclusion': 'Reject null: houses close to a school or hospital are more expensive' if p_value < alpha else 'Fail to reject null: no significant price difference'
    }

results = test_house_value_by_proximity(housing_df)
print(results)
```
